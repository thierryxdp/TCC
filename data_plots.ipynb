```python
def median(data):
    sorted_data = sorted(data)
    n = len(sorted_data)
    if n % 2 == 0:
        # For even number of observations, average of two middle values
        middle = n // 2
        median_value = (sorted_data[middle - 1] + sorted_data[middle]) / 2
    else:
        # For odd number of observations, middle value
        median_value = sorted_data[n // 2]
    return median_value
```

```python
def seconds_to_hms(seconds):
    hours = round(seconds // 3600)
    minutes = round((seconds % 3600) // 60)
    seconds = round(seconds % 60)
    print(f"Total time: {hours} hours, {minutes} minutes, and {seconds} seconds.")
```
